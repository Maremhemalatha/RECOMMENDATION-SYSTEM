```python
import pandas as pd
import numpy as np
from surprise import Dataset, Reader, SVD
from surprise.model_selection import cross_validate, train_test_split
from surprise import accuracy
import matplotlib.pyplot as plt
reader = Reader(line_format='user item rating timestamp', sep='\t')
data = Dataset.load_from_file('C:\\Users\\marem\\Downloads\\task4\\ml-100k\\u.data', reader=reader)
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
#Use SVD (Singular Value Decomposition)
model = SVD()
model.fit(trainset)
predictions = model.test(testset)
# Evaluate performance
rmse = accuracy.rmse(predictions)
mae = accuracy.mae(predictions)
# Predict rating for specific user and item
user_id = str(196)
item_id = str(302)
prediction = model.predict(user_id, item_id)
print(f"Predicted rating by user {user_id} for movie {item_id} is {prediction.est:.2f}")
all_items = trainset.all_items()
all_items = [trainset.to_raw_iid(i) for i in all_items]
# Recommend top N items not yet rated by the user
def get_top_n_recommendations(user_id, n=5):
    user_id = str(user_id)
    rated_items = set([j[0] for j in trainset.ur[trainset.to_inner_uid(user_id)]])
    predictions = []
    for item_id in all_items:
        if item_id not in rated_items:
            pred = model.predict(user_id, item_id)
            predictions.append((item_id, pred.est))
    top_n = sorted(predictions, key=lambda x: x[1], reverse=True)[:n]
    return top_n
top_recs = get_top_n_recommendations(196, n=5)
print("Top 5 recommendations for User 196:")
for item_id, rating in top_recs:
    print(f"Item ID: {item_id}, Predicted Rating: {rating:.2f}")
pred_ratings = [pred.est for pred in predictions]
plt.hist(pred_ratings, bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Predicted Ratings")
plt.xlabel("Predicted Rating")
plt.ylabel("Frequency(no.of user-item pairs received that rating)")
plt.show()
```
